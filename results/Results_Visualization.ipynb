```python
#Text File Cleaning Step

def clean_data(filename):
    with open(filename, 'r') as infile:
        lines = [line.strip() for line in infile if line.strip().startswith(('real', 'user', 'sys'))]

    with open(filename, 'w') as outfile:
        for line in lines:
            outfile.write(line + '\n')
```

```python
import matplotlib.pyplot as plt

clean_data('dataOpenMPSeq.txt')

def time_str_to_seconds(tstr):
    min_part, sec_part = tstr.strip().split('m')
    return int(min_part) * 60 + float(sec_part[:-1])

real_times = []
user_times = []
sys_times = []

with open('dataOpenMPSeq.txt', 'r') as file:
    lines = [line.strip() for line in file if line.strip()]
    for i in range(0, len(lines), 3):
        real_times.append(time_str_to_seconds(lines[i].split()[1]))
        user_times.append(time_str_to_seconds(lines[i+1].split()[1]))
        sys_times.append(time_str_to_seconds(lines[i+2].split()[1]))

x = list(range(1, len(real_times) + 1))

plt.figure(figsize=(12, 6))
plt.plot(x, real_times, label='Real Time', marker='o')
plt.plot(x, user_times, label='User Time', marker='x')
plt.plot(x, sys_times, label='Sys Time', marker='s')
plt.xticks(x)
plt.xlabel("Number of Iterations")
plt.ylabel("Time (seconds)")
plt.title("Performance Analysis: OpenMP Sequential Test")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
